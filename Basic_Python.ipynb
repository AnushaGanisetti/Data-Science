```python
#checking odd or even using a function
def check_even_odd(num):
    if num % 2 == 0:
        return "Even Number"
    else:
        return "Odd Number"

n = int(input("Enter a number: "))
result = check_even_odd(n)
print(result)
```

```python
#positional arguments
def sub(x,y):
  print(x-y)
sub(20,10)
```

```python
#keyword arguments
def products(a,b):
  print("The products are: ",a,b)
products(a="Laptop",b="Computer")
```

```python
#Default arguments
def details(name,age=21):
  print(name,age)
details("Anu",18)
```

```python
#variable length arguments
def info(name,*skills):
  print("Name: ",name)
  print("Skills: ",skills)
info("Anu","python","java","SQL")
```

```python
#lambda function
a=lambda x:x*x
s=a(3)
print(s)
```

```python
#lambda function using map
ages=[12,18,13,17]
ages_after_two_years=map(lambda x:x+2,ages)
print(list(ages_after_two_years))
```

```python
#lambda function using filter
ages=[12,28,13,26,30,19]
ages_eligible_for_vote=filter(lambda x:x>18,ages)
print(list(ages_eligible_for_vote))
```

```python
#lambda function using reduce
from functools import reduce
ages=[12,28,13,26,30,19]
total_of_ages=reduce(lambda x,y:x+y,ages)
print(total_of_ages)
```
